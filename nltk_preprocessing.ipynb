```python
""" Tokenizing the sentences to split the sentences to help the model easy
to find the relation and then tokenizing the word in the sentences to see how
each word related to other and see more context"""
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
paragraph = """
... Muad'Dib learned rapidly because his first training was in how to learn.
... And the first lesson of all was the basic trust that he could learn.
... It's shocking to find how many people do not believe they can learn,
... and how many more believe learning to be difficult."""
```

```python
example_sent = sent_tokenize(paragraph)
example_sent
```

```python
example_word = word_tokenize(paragraph)
example_word
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
for i in example_word:
  for word in i:
    if word not in set(stopwords.words('english')):
      continue
print(stopwords.words('english'))
```

```python
filtered_list = []
stop_words = set(stopwords.words('english'))
for word in example_word:
  if word.casefold() not in stop_words:
    filtered_list.append(word)
filtered_list
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed = [
    stemmer.stem(word)
    for word in filtered_list
    if word not in stop_words
]

print(stemmed)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
nltk.pos_tag(filtered_list)
```

```python
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

# Ensure wordnet is downloaded (already done, but good practice to include if cell is standalone)
nltk.download('wordnet', quiet=True)

# Function to convert NLTK's POS tags to WordNet's POS tags
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Default to noun if no clear tag found

lemmatizer = WordNetLemmatizer()
lemmatized_words = []

# Get POS tags for filtered_list
# This also downloads 'averaged_perceptron_tagger' if not already present
pos_tags_filtered_list = nltk.pos_tag(filtered_list)

for word, tag in pos_tags_filtered_list:
    # Lemmatize word using its WordNet POS tag
    wordnet_pos = get_wordnet_pos(tag)
    lemmatized_words.append(lemmatizer.lemmatize(word, wordnet_pos))

print(lemmatized_words)
```

```python
""" TF-IDF resolves the issue of BoW and extracts the feature from documents
    BoW we exatract common words which repeats in document has high priority
    but when it comes to documents BoW has limitation, So Tf - Idf resolved that
    Tf - Idf core idea was frequent in document but rare across the documents """

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(filtered_list)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```
