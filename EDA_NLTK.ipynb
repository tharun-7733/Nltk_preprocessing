```python
import nltk
nltk.download('book')
from nltk.book import *
```

The Book is a library where it contains corpora covering everything from novels hosted by Project Gutenberg to inaugural speeches by presidents of the United States.

```python
""" Concordance is alphabetical index where it shows the all occurences of words
along with the surrounding words. This is very useful in Exploratory Data Analysis
and also to understand the context of corpus. """

text1.concordance('man')
```

```python
""" Dispersion plot is useful in analyzing the occurences of particular word
in different corpora and where it appears """

text1.dispersion_plot(
    ["man", "woman", "lady", "girl", "gal", "gentleman", "boy", "guy"]
)
```

```python
# Frequency distribution you can check which words shows up the most frequently in your text
from nltk import FreqDist
freq_distribution = FreqDist(text1)
print(freq_distribution)
```

```python
freq_distribution.most_common(30)
```

```python
from nltk.corpus import stopwords
meaningFull_words = [
    word for word in text1 if word.casefold() not in set(stopwords.words('english')) and word.isalpha()
]
freq_dist_meaning = FreqDist(meaningFull_words)
freq_dist_meaning.most_common(30)
```

```python
freq_dist_meaning.plot(30, cumulative=True)
```
